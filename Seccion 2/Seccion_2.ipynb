```python
gpa = data['GPA']
gpaMenosMedia = []
mediaGPA = 3.47

for i in range(len(gpa)):
  gpaMenosMedia.append(mediaGPA - gpa[i])

ax=sb.distplot(gpaMenosMedia, bins = 10)
```

```python
gpaMenosMedia.sort()
#valores maximos y minimos
min = gpaMenosMedia[0]
max = gpaMenosMedia[201]
 
#intervalo total de 2
intervalo_total = max - min

#intervalos de 0.20
frecuencias = []
probabilidades = []
for n in range(11): 
  frecuencia=0
  for i in range(202):
      if (gpaMenosMedia[i]) >= (min + n*0.20) and (gpaMenosMedia[i] < (min + (n+1)*0.20)):
        frecuencia +=1
  frecuencias.append(frecuencia)
# calculo de probabilidades
for i in range(11):
  probabilidad = frecuencias[i]/202
  probabilidades.append(probabilidad)

total = 0
for i in range(11):
  total += probabilidades[i]
```

```python
valores= []
marcas=[]
valor_esperado_1 = 0

for n in range(10): 
  marca1 = (min + n*0.20)
  marca2=(min + (n+1)*0.20)
  marca = (marca1 + marca2)/2
  marcas.append(marca)
marcas.append(max)

for i in range(11):
  valores.append(marcas[i]*probabilidades[i])

for i in range(11):
  valor_esperado_1 += valores[i]
print("Valor Esperado variable GPA: ")
valor_esperado_1
```

```python
#Varianza
#E(x^2)
esperado_cuadrados1 = 0
varianza1 = 0 
cuadrados1=[]
frec_cuadrados1=[]
prob_cuadrados1=[]

for i in range(202):
  cuadrado = (gpaMenosMedia[i])**2
  cuadrados1.append(cuadrado)

cuadrados1.sort()
cuadrados1_min= cuadrados1[0]
cuadrados1_max= cuadrados1[201]
intervalo = cuadrados1_max - cuadrados1_min

#intervalos de 0.216
for n in range(11): 
  frecuencia=0
  for i in range(202):
      if (cuadrados1[i]) >= (cuadrados1_min + n*0.216) and (cuadrados1[i]) < (cuadrados1_min + (n+1)*0.216):
        frecuencia +=1
  frec_cuadrados1.append(frecuencia)
# calculo de probabilidades
for i in range(11):
  probabilidad = frec_cuadrados1[i]/202
  prob_cuadrados1.append(probabilidad)

marcas_cuadrados1=[] 

for n in range(10): 
  marca1 = (cuadrados1_min + n*0.216)
  marca2=(cuadrados1_min + (n+1)*0.216)
  marca = (marca1 + marca2)/2
  marcas_cuadrados1.append(marca)
marcas_cuadrados1.append(cuadrados1_max)

for i in range(11):
  esperado_cuadrados1 += (marcas_cuadrados1[i]*prob_cuadrados1[i])

#E(x^2)-ux^2 
varianza1 = esperado_cuadrados1 - (valor_esperado_1**2)
print("Varianza variable GPA:" + str(varianza1))
desvesta = math.sqrt(varianza1)
print("Desviacion Estandar:" + str(desvesta))
```

```python
#AGRUPO DATOS
marcaClase = [19,21,23,25,29.5]
agrupamiento = []
menor20 = 0
menor21 = 0
menor23 = 0
menor25 = 0
menor29 =0

for i in range(202):
  if(dataEdad[i]<20):
    menor20+=1
  elif(dataEdad[i]>=20 and dataEdad[i]<22):
    menor21+=1
  elif(dataEdad[i]>=22 and dataEdad[i]<24):
    menor23+=1
  elif(dataEdad[i]>=24 and dataEdad[i]<26):
    menor25+=1
  elif(dataEdad[i]>=26):
    menor29+=1

agrupamiento.append(menor20)
agrupamiento.append(menor21)
agrupamiento.append(menor23)
agrupamiento.append(menor25)
agrupamiento.append(menor29)

#OBTENGO PROBABILIDADES
probabilidades =[]
for i in range(len(agrupamiento)):
  probabilidades.append(agrupamiento[i]/202)

#OBTENGO ESPERANZA DE VALORES AGRUPADOS
lambdaF = 0
for i in range(len(marcaClase)):
  lambdaF += marcaClase[i]*probabilidades[i]
print("LAMBDA DE MIS DATOS AGRUPADOS:" + str(lambdaF))
```
